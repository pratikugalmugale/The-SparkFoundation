## Toss Decision

```python
teams =  data['toss_winner'].unique()
decision_making = pd.DataFrame([],columns=['Toss Winner','Decision','Times'])
for id,element in enumerate(teams):
  temp_bat = data[(data['toss_winner']==element) & (data['toss_decision']=='bat')]
  temp_fields = data[(data['toss_winner']==element) & (data['toss_decision']=='field')]

  decision_making = decision_making.append({'Toss Winner':element,'Decision':'bat','Times':temp_bat['toss_winner'].count()},ignore_index=True)
  decision_making = decision_making.append({'Toss Winner':element,'Decision':'fields','Times':temp_fields['toss_winner'].count()},ignore_index=True)
```

```python
decision_making.head(10)
```

```python
sns.catplot(x='Toss Winner',y='Times',hue='Decision',data=decision_making,kind='bar',height=5,aspect=2)
plt.xticks(rotation=90)
plt.title('Toss deciosion per team')
plt.xlabel('IPL teams')
plt.ylabel('Toss Decision')
plt.show()
```

## Famous Venue

## Wins at Home ground

```python
venue  = data.groupby(['venue','winner']).count()
venue = venue.iloc[:-1,:1].reset_index()
venue = venue.pivot('venue','winner')
venue.fillna(0,inplace=True)
venue['total'] = venue.sum(axis=1,skipna=False)
venue = venue.sort_values(by=['total'],ascending=False).head(9)
venue.drop(['total'],axis=1,inplace=True)

sns.set_style('dark')
plt.figure(figsize=(16,9))
sns.heatmap(data=venue,annot=True,cmap='rocket')
plt.xlabel('Team',size=15)
plt.ylabel('Ground',size=15)
plt.xticks(size=12)
plt.yticks(size=12)
plt.title('Team performance (Grounds)',size=20)
plt.show()
```
